How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
import calendar 

msteps_by_month = df.groupby('Months_encoded', as_index=False)[['Months_encoded','Steps']].median()
plt.figure(figsize=(12,4))
fig, axs = plt.subplots(1,2)
plt.sca(axs[0])
plt.plot(msteps_by_month.Months_encoded,msteps_by_month.Steps)
#plt.xticks(np.arange(12), calendar.month_name[1:13], rotation=20)
plt.xticks(np.arange(12),['january','','','','','','july','','','','','december'], rotation=20)
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
msteps_by_ww = df.groupby('Work_or_Weekend', as_index=False)[['Work_or_Weekend','Steps']].median()
#plt.figure(figsize=(12,4))
plt.sca(axs[1])
plt.bar(msteps_by_ww.Work_or_Weekend, msteps_by_ww.Steps)
plt.xticks(np.arange(2), ['work','weekend'], rotation=20)
plt.savefig('../figures/challenge2.png')
```
